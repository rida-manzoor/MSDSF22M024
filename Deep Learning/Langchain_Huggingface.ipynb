# HuggingFace Endpoint
## How to access Huggingface model with Api?

```python
from langchain_huggingface import HuggingFaceEndpoint
```

```python
## Environment secret key
from google.colab import userdata
sec_key = userdata.get('HUGGINGFACEHUB_API_TOKEN')
print(sec_key)
```

```python
import os
os.environ['HUGGINGFACE_API_TOKEN'] = sec_key
```

```python
repo_id = 'mistralai/Mistral-7B-Instruct-v0.2'
llm = HuggingFaceEndpoint(repo_id=repo_id, max_length=128,temperature=0.5,token=sec_key)
```

```python
llm
```

```python
llm.invoke('What is cat?')
```

```python
from langchain import PromptTemplate, LLMChain

question = "What is cat?"
template = """Question: {question}

Answer: Let's think step by step."""
prompt = PromptTemplate(template=template, input_variables=["question"])
```

```python
print(prompt)
```

```python
llm_chain = LLMChain(llm=llm, prompt=prompt)
print(llm_chain.invoke(question))
```

# HuggingFace Pipeline

```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
```

```python
model_id = "gpt2"
model = AutoModelForCausalLM.from_pretrained(model_id)
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
pipeline = pipeline("text-generation", model=model_id,
                    tokenizer=tokenizer, max_new_tokens=100)
hf = HuggingFacePipeline(pipeline=pipeline)
```

```python
hf
```

```python
hf.invoke("Cat is a animat")
```

```python
# Use HuggingfacePipeline with GPU
gpu_llm = HuggingFacePipeline.from_model_id(model_id=model_id,
                                            task="text-generation",
                                            device=0,
                                            pipeline_kwargs={"max_new_tokens": 100})
```

```python
from langchain import PromptTemplate, LLMChain

question = "What is cat?"
template = """Question: {question}

Answer: Let's think step by step."""
prompt = PromptTemplate(template=template, input_variables=["question"])
```

```python
chain = prompt | gpu_llm
```

```python
question = "What is cat?"
chain.invoke({"question": question})
```
